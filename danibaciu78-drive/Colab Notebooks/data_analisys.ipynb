```python
print(os.listdir('./data'))
import pandas as pd
import numpy as np
```

```python
def get_all_data_from_filename(full_filename):
    df = pd.read_csv(full_filename, delimiter='\t', header=None, names=['iid','src','native_speaker','original','dest','text','direct','label'])
    return df
```

```python
def get_text_and_label(df):
    return df.text.values[1:], df.label.values[1:]
```

```python
data_path = './data'
en_de = 'en_de'
en_es = 'en_es'

infos = []
```

```python
def get_info(smth):
  for filename in os.listdir(data_path):
    if smth in filename :
      all_data = get_all_data_from_filename(os.path.join(data_path,filename))
      propositions, labels = get_text_and_label(all_data)
      infos.append(['german' if smth == en_de else 'spain', filename, len(propositions), np.count_nonzero(labels == '1'), np.count_nonzero(labels == '0')])
      print(f"filename = {filename}, size = {len(propositions)} label1 = {np.count_nonzero(labels == '1')} label0 = {np.count_nonzero(labels == '0')}")
```

```python
get_info(en_de)
get_info(en_es)
```
